```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
dff=scaler.fit_transform(df.drop('TARGET CLASS',axis=1))

dff
```

```python
df_feat=pd.DataFrame(dff,columns=df.columns[:-1])
df_feat.head()
```

# Selecting our x and y features

```python
x=df_feat
y=df['TARGET CLASS']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

# K NEAREST NEIGHBORS

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

# PREDICTIONS AND EVALUATIONS

```python
predictions=knn.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics  import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))

print('\n')

print(confusion_matrix(y_test,predictions))
```

# ELBOW METHOD

```python
error_rate=[]

for  i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i=knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker="o",markerfacecolor='black',markersize=10)
plt.title('Error_rate vs K-value')
plt.xlabel('k-values')
plt.ylabel('Error_rate')
```

```python
knn=KNeighborsClassifier(n_neighbors=23)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
print(classification_report(y_test,pred))

print('\n')


print(confusion_matrix(y_test,pred))
```
